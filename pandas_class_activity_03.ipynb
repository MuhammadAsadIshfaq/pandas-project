# Task:4

```python
df['Car Age'] = pd.to_datetime('now').year - df['Prod. year']
```

```python
mileage_bins = [0, 50000, 100000, float('inf')]
mileage_labels = ['Low Mileage', 'Medium Mileage', 'High Mileage']

df['Mileage Category'] = pd.cut(df['Mileage'], bins=mileage_bins, labels=mileage_labels)
```

```python
df['Fuel type'] = pd.to_numeric(df['Fuel type'], errors='coerce')

df['Fuel Efficiency'] = df['Mileage'] / df['Fuel type']
```

```python
df.head()
```

```python
df['Luxury Car'] = (df['Leather interior'] == 'Yes') & (df['Airbags'] > 0)
df.head()
```

```python
bins = [0, 1.5, 3.0, float('inf')]
labels = ['Small', 'Medium', 'Large']

df['Engine Volume Category'] = pd.cut(df['Engine volume'], bins=bins, labels=labels)
```

```python
color_popularity = df['Color'].value_counts(normalize=True)
df['Color Popularity'] = df['Color'].map(color_popularity)
```

```python
df.head()
```

```python
market_share = df['Manufacturer'].value_counts(normalize=True)
df['Market Share'] = df['Manufacturer'].map(market_share)
```
