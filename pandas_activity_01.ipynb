```python
data = {
    'id': ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10'],
    'Name': ['Laptop', 'Smartphone', 'Television', 'Refrigerator', 'Washing Machine',
                     'Air Conditioner', 'Microwave Oven', 'Blender', 'Vacuum Cleaner', 'Coffee Maker'],
    'Price': [1200, 800, 1500, 1000, 1200, 1800, 400, 50, 200, 100],
    'Quantity': [50, 80, 30, 45, 35, 20, 60, 150, 40, 120],
    'Discount': [5, 10, 8, 6, 7, 12, 4, 2, 5, 3],
    'Rating': [4.2, 4.5, 4.1, 4.3, 4.4, 4.0, 4.6, 4.8, 4.2, 4.7]
}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.set_index('id', inplace=True)
```

```python
df
```

```python
df['Total Revenue'] = (df['Price'] * df['Quantity'] * (1 - df['Discount'] / 100))
df
```

```python
df['Total Revenue'].max()
```

```python
ratingg = df.sort_values(by='Rating', ascending=False)
ratingg
```

```python
df[df['Rating'] > 4.2]
```
